### 3.1TensorDecompositionModels
### *3.1 Tensor Decomposition Models*

Các mô hình trong họ này giải thích LP là một nhiệm vụ [tensor decomposition-28](../papers/28_Tensor_Decompositions_and.pdf). Các mô hình của họ này hoàn toàn coi KG là ma trận kề 3D (nghĩa là một **3-way tensor)**), chỉ có thể quan sát được một phần do sự không hoàn chỉnh của KG. **tensor** được phân tách thành một tổ hợp (ví dụ: một tích đa tuyến tính-*multi-linear product*) của các vectơ có chiều thấp *low-dimensional vectors*: các vectơ như vậy được sử dụng làm nhúng-*embeddings* cho các thực thể và quan hệ. Ý tưởng cốt lõi là, với điều kiện là mô hình không *overfit* trên tập huấn luyện, các *embeddings* được học sẽ có thể khái quát hóa và liên kết các giá trị cao với các sự kiện-facts không xác thực trong ma trận kề của đồ thị. Trong thực tế, điểm số của mỗi facts được tính toán hoạt động kết hợp trên các *embeddings* cụ thể liên quan đến facts đó; các *embeddings* được học như bình thường bằng cách tối ưu hóa hàm cho điểm cho tất cả các training facts. Các mô hình có xu hướng sử dụng ít hoặc không có tham số chia sẻ-*shared parameters* nào cả; điều này làm cho chúng đặc biệt nhẹ và dễ huấn luyện.

#### BilinearModels
### 3.1.1 Bilinear Models

Với head embedding $h \in \mathbb{R}^d$ và tail embedding $t \in \mathbb{R}^d$, các mô hình này biểu thị mối quan hệ nhúng-*relation embedding* dưới dạng ma trận hai chiều $r \in \mathbb{R}^{d \times d}$. Hàm chấm điểm sau đó được tính là một tích song tuyến **bilinear product**:
$$
\phi(h, r, t) = h \times r \times t
$$
trong đó ký hiệu $\times$ biểu thị **matrix product-nhân 2 ma trận**. Các mô hình này thường khác nhau bằng cách đưa ra các ràng buộc bổ sung cụ thể trên các nhúng mà chúng học được. Đối với nhóm này, trong phân tích so sánh của chúng tôi, chúng tôi bao gồm các mô hình đại diện sau:

[DistMult-71](../papers/71_Embedding_Entities_and_Relations_for_Learning_and_Inference_in_Knowledge_Bases.pdf)

buộc tất cả các nhúng quan hệ phải là ma trận đường chéo, điều này luôn làm giảm không gian của các tham số cần học, dẫn đến một mô hình dễ huấn luyện hơn nhiều. Mặt khác, điều này làm cho hàm tính điểm giao hoán, với $\phi(h, r, t) = \phi(t, r, s)$, có nghĩa là coi tất cả các quan hệ là đối xứng. Mặc dù vậy, điều này đã được chứng minh rằng, khi được điều chỉnh cẩn thận, DistMult vẫn có thể đạt được hiệu suất tốt [state-of-the-art performance-26](../papers/26_Knowledge_Base_Completion_Bas_lines_Strike_Back.pdf).

[ComplEx-61](../papers/61_Complex_Embeddings_for_Simple_Link_Prediction.pdf)

tương tự như DistMult, buộc mỗi quan hệ nhúng phải là một ma trận đường chéo, nhưng mở rộng công thức như vậy trong không gian phức(số phức): $h \in \mathbb{C}^d, \ t \in \mathbb{C}^d, \ r \in \mathbb{C}^{d \times d}$. Trong không gian phức, **bilinear product** trở thành **Hermitian product**, thay vì $t$ gốc, $\bar{t}$ chuyển vị liên hợp-**conjugate-transpose** của nó được sử dụng. Điều này sẽ vô hiệu hóa tính giao hoán ở trên được đề cập cho hàmg cho điểm, cho phép ComplEx cũng mô hình thành công các mối quan hệ bất đối xứng.

[ANALOGY-35](../papers/35_Analogical_Inference_for_Multi-relational_Embeddings.pdf)

nhằm mục đích mô hình hóa lý luận tương tự, đó là chìa khóa cho bất kỳ loại cảm ứng tri thức nào. Nó sử dụng hàm tính điểm song tuyến tổng quát nhưng thêm hai ràng buộc chính được lấy cảm hứng từ các cấu trúc tương tự: $\mathbf{r}$ phải là một ma trận bình thường: $\mathbf{r \ r}^T = \mathbf{r}^T \mathbf{r}$; đối với mỗi cặp quan hệ $r_1, r_2$, thành phần của chúng phải có tính giao hoán: $r1 \circ r2 = r_2 \circ r_1$. Các tác giả chứng minh rằng ma trận bình thường có thể được sử dụng thành công để mô hình hóa các mối quan hệ bất đối xứng.

[SimplE-27](../papers/27_SimplE_Embedding_for_Link_Prediction_in_Knowledge.pdf)

buộc các quan hệ *embeddings* (**relation embeddings**) là ma trận đường chéo, tương tự như **DistMult**, nhưng mở rộng nó bằng cách (i) liên kết với mỗi thực thể $e$ hai *embeddings, $eh$ và $et$, tùy thuộc vào việc $e$ được sử dụng làm **head** hay **tail**; (ii) liên kết với mỗi quan hệ $r$ hai ma trận đường chéo riêng biệt, $r$ và $r_{-1}$, biểu thị mối quan hệ theo hướng thông thường và nghịch đảo của nó. Điểm số của một sự kiện - **fact** được tính trung bình cho điểm số song tuyến tính **bilinear scores** thông thường **regular fact** và phiên bản nghịch đảo của nó. Nó đã được chứng minh rằng SimplE hoàn toàn biểu cảm, và do đó không giống như DistMult, nó có thể mô hình hóa các mối quan hệ bất đối xứng.

### *3.1.2 Non-bilinear Models*

Các mô hình kết hợp các phần nhúng **head**, **relation and tail embeddings** của *kết hợp* bằng cách sử dụng các công thức khác nhau từ tích song tuyến nghiêm ngặt **strictly bilinear product**.

[HolE-46](../papers/46_Holographic_Embeddings_of_Knowledge_Graphs.pdf)

Thay vì sử dụng các tích song tuyến tính, hãy tính tương quan vòng tròn **circular correlation** (ký hiệu là $\star$ Trong Bảng 1) giữa các phần nhúng của các thực thể đầu và đuôi **embeddings of head and tail entities**; sau đó, nó thực hiện phép nhân ma trận với việc nhúng quan hệ **relation embedding**. Lưu ý rằng trong mô hình này, các relation embedding có hình dạng giống như embeddings entities. Các tác giả đã chỉ ra rằng mối tương quan vòng tròn **circular correlation** có thể được xem là sự nén của tích ma trận đầy đủ: điều này làm cho HolE đỡ tốn chi phí hơn với mô hình song tuyến không bị giới hạn về cả độ phức tạp của thời gian và không gian.

[TuckER-6](../papers/6_TuckER_Tensor_Factorization_for_Knowledge_Graph_Completion.pdf)

dựa vào  Tucker decomposition [21], yếu tố này một tensor thành một tập các vectơ và lõi $\textbf{W}$ nhỏ hơn. Mô hình TuckER học $\textbf{W}$ cùng với các nhúng KG *KG embeddings*. Trên thực tế, việc học các tham số được chia sẻ trên toàn cục là khá hiếm trong các **Matrix Factorization Models**; các tác giả giải thích rằng $\textbf{W}$ có thể được xem như là một nhóm chung của ma trận quan hệ nguyên mẫu, được kết hợp theo cách khác nhau cho mỗi mối quan hệ tùy thuộc vào sự nhúng của nó. Trong TuckER, kích thước của các nhúng thực thể và quan hệ là độc lập với nhau, với các nhúng thực thể $e \in \mathbb{R}^{d_e}$ và nhúng nhúng quan hệ $r \in \mathbb{R}^{d_r}$. Hình dạng-Shape của $W$ phụ thuộc vào kích thước của các thực thể và quan hệ, với $W \in \mathbb{R}^{d_e \times d_r \times d_e}$. Trong Bảng 1, chúng tôi biểu thị bằng $\times i$ *tensor product* dọc theo chế độ i được sử dụng bởi TuckER.

## *3.2 Geometric Models*
**Geometric Models** diễn giải các mối quan hệ như các biến đổi hình học trong không gian ẩn. Với một fact, phần **head** nhúng trải qua một phép biến đổi không gian $\tau$ sử dụng các giá trị của phần nhúng quan hệ làm tham số **relation embedding as parameters**. Điểm thực tế là khoảng cách giữa vectơ kết quả và vectơ **tail**; một offset như vậy được tính bằng cách sử dụng hàm khoảng cách $\delta$ (ví dụ: L1 của L2 norm).
$$
\phi(h, r, t) = \delta(\tau(h, t), r)
$$

Tùy thuộc vào dạng phân tích của $\tau$, các mô hình hình học có thể chia sẻ các điểm tương đồng với các **Tensor Decomposition models**, nhưng trong các trường hợp này, các mô hình hình học thường cần phải thực thi các ràng buộc bổ sung để thực hiện phép biến đổi không gian hợp lệ của chúng. Ví dụ, phép quay được vận hành bởi mô hình **RotatE** có thể được định dạng như một tích ma trận, nhưng ma trận xoay sẽ cần phải có đường chéo và có các phần tử với mô đun 1-modulus 1.

Giống như với **Matrix Factorization Models**, các hệ thống này thường tránh các tham số được chia sẻ, chạy lan truyền ngược trực tiếp trên các **embeddings**. Chúng tôi xác định ba nhóm trong họ này: (i) *Pure Translational Models* (Mô hình tịnh tiến thuần túy) (ii) *Translational
Models with Additional Embeddings* (Mô hình tịnh tiến với các nhúng bổ sung) và (iii) *Roto-translational models* (Mô hình dịch mã Roto).

### 3.2.1 Pure Translational Models
Các mô hình này diễn giải mỗi quan hệ là một bản dịch *translation* trong không gian ẩn: việc nhúng quan hệ chỉ được thêm vào phần **head embedding** và hy vọng sẽ dừng lại ở vị trí gần với phần *tail embedding*. do đó các mô hình này đại diện cho các thực thể và quan hệ như các vectơ một chiều có cùng độ dài.

[TransE-8](../papers/8_Translating_Embeddings_for_Modeling.pdf)

là mô hình LP đầu tiên đề xuất một diễn giải hình học về không gian ẩn, phần lớn lấy cảm hứng từ khả năng quan sát trong các vectơ [Word2vec-39](../papers/39_Efficient_Estimation_of_Word_Representations_in_Vector_Space.pdf) để nắm bắt các mối quan hệ giữa các từ dưới dạng dịch giữa các phần nhúng của chúng. TransE thực thi điều này một cách rõ ràng, yêu cầu ** tail embedding** nằm gần với tổng của các phần **head and relation embeddings**, theo hàm khoảng cách đã chọn. Do bản chất của dịch thuật, TransE không thể xử lý chính xác các mối quan hệ một-nhiều và nhiều-một, cũng như các mối quan hệ đối xứng và bắc cầu.

## *3.3 Deep Learning Models*

Deep Learning sử dụng những mạng neural học sâu để thực hiệ nhiệm vụ LP.
Mạng neural (Neural Networks) học các tham số giống như trọng số và bias, mà chúng có thể kết hợp với dữ liệu đầu vào để
nhận dạng những mẫu có ý nghĩa. Mạng neural học sâu thì thường tổ chức các tham số thành nhiều layer khác nhau,
thường xen kẽ với các hàm kích hoạt phi tuyến tính (non-linear activation function).
Cho đến nay, có rất nhiều layer đã được phát triển, dùng để tính toán vào rất nhiều dữ liệu đầu vào khác nhua
Ví dụ như Dense layer, sẽ kết hợp cả dữ liệu đầu vào X với trọng số W và thêm một bias B:
W X X = B
Ở những layer nâng cao để tính toán phức tạp, ví dụ như convolutional layer (lớp tính chập). Có thể học convolution kernel
(Nhân tích chập) để dùng trong dữ liệu đầu vào hoặc recurrent layers (lớp hồi qui) để xử lý các dữ liệu tuần tự trong một recursive fashion

Trong lĩnh vực LP, KG nhúng thì thường được học chung với trong số và bias của layer; Chúng chia sẻ những tham số làm cho những
mô hình này ý nghĩa hơn, nhưng có khả năng năng hơn và khó huấn luyện hơn, và dễ bị overfitting hơn
Bài báo chia ra 3 nhóm dự vào cấu trúc của mạng neural :
- Convolutional Neural Networks (Mạng neural tích chập)
- Capsule Neural Networks (Mạng bao bọc)
- Recurrent Neural network (Mạng hồi qui)

### 3.3.1 Convolutional Neural Networks (Mạng neural tích chập)

Đây là những mô hình sử dụng một hoặc nhiều lớp tích chập [[33]](https://github.com/hmthanh/thesis/blob/master/papers/33.%20Gradient-based%20learning%20applied%20to%20document%20recognition.pdf) :
Mỗi lớp thực hiện tich chập trên dữ liệu đầu vào trên những bộ lọc chiều thấp (low-dimensional filter) w. Kết quả là một
bản đồ các đặc trưng (feature map) để thêm các lớp mạng dày (dense layer) để mà tính toán các điểm số kết quả.

* **ConvE** [[11]](https://github.com/hmthanh/thesis/blob/master/papers/11_Convolutional_2D_Knowledge_Graph_Embeddings.pdf) 
thể hiện những thực thể và mối liên hệ như một chiều kích thước d đã nhúng. Khi tính toán điểm số của sự kiện, nó ghép (concatenates) và thay dổi kích thước nút gốc và liên kết nhúng **h**, **r** vào một đầu vào duy nhất [**h**, **r**] ;
Và đặt tên nó là miền $ d_m \times d_n $. Đầu vào này qua một lớp tích chập với một tập hợp $ \omega $ của $ m \times n $ bộ lọc và sau đó đi qua một lớp Dense layer với d neuron và một tập hợp trong số $ W $. Kết quả cuối cùng kết hợp của với nút nhúng kết quả. $ t $ sử dụng dot product. kết quả là điểm số sự thật. Khi sử dụng toàn bộ ma trận của thực thể nhúng thay vì nhúng một thực thể mục tiêu $ t $ , kiến trúc này có thể coi như là một phân loại với $ | \varepsilon | $ lớp

* **ConvKB** [[42]](https://github.com/hmthanh/thesis/blob/master/papers/42_A_Novel_Embedding_Model_for_Knowledge_Base_Completion_Based_on_Convolutional_Neural_Network.pdf) mô hình sử dụng thực thể và liên hệ cùng kích thước nhúng miền một chiều (same-sized one-dimensional embeddings). Khác với Conv2E, với bất kỳ sự kiện $ \langle h, r, t \rangle $. nó ghép tất cả những cái nhúng **h**, **r** và **t** vào một $ d \times 3 $ ma trận đầu vào $ [hr, r, t] $ . Đầu vào này truyền qua một lớp tích chập với một tập hợp $ \omega $ cuar $ T $ bộ lọc kích thước $ 1 \times 3 $, kết quả là một bàn đồ đặc trưng $ T  \times 3 $. Mâ trận dặc trưng này đi qua một lớp Dense layer với chỉ một neuron và một trọng số $ W , kết quả là điểm số của kiện. Cấu trúc này có thể coi như là phân loại nhị phân, gọi ra xắc xuất mà sự đúng.

* **ConvR** [25](https://github.com/hmthanh/thesis/blob/master/papers/25.%20Adaptive%20Convolution%20for%20Multi-Relational%20Learning.pdf) thể hiện thực thể và liên hệ nhúng như là một miền vector của những miền khác $ d_e $ và $ d_r $.
CHo bất kỳ sự kiện $ \langle h, r, t \rangle $, **h** đầu tiên được thay đổi kích thước về mâ trận có kích thước $ d_{e_m}, d_{e_n} $ với $ d_{e_m} \times d_{e_n} = d_e $ . **r** sau đó được thay đổi kích thước và chia thành tập hợp $ \omega_r $ của tập hợp $ T $ bộ lọc, với mỗi bộ lọc có kích thước $ m \times n $ . Những bộ lọc này sau đó được sử dụng để chạy tích chập trên $ h $; sau đó được tích chập kép (adaptive convolution) (https://www.youtube.com/watch?v=gsQZbHuR64o) với những bộ lọc quan hệ đặc biệt (relation-specific filters)
Kết quả bản đồ dặc trưng thông qua Dense Layer với trọng số $ W $, như là ConvE, điểm sự kiện thì bao gồm kết quả đầu ra neural với tail embedding $ t $ sử dụng dot product

### 3.3.2 Capsule Neural Networks

Capsulate Networks (CapsNets) là hợp thành từ những nhóm neurons, được gọi là capsules (bao bọc) mà mã háo những đặc trưng đặc biệt của đầu vào, ví dụ như biểu diễn những chi tiết đối tượng của một hình ảnh. CapsNets được thiết kế nhận dạng những đặc trưng mà không mất thông tin không gian mà mạng tích chập làm. Mỗi một lớp bọc gửi đầu ra của chúng những cái có thứ tự ưu tiên cao hơn, với kết nối được quyết định bởi quá trình định tuyến động. Xắc xuất của lớp bọc đặc trưng được được phát hiện theo chiều dài của vector đầu ra.

* **CapsE** [ [43]](https://github.com/hmthanh/thesis/blob/master/papers/43_A_Capsule_Network-based_Embedding_Model_for_Knowledge_Graph_Completion_and_Search_Personalization.pdf) nhúng những thực thể và liên hệ vào một vector một miền kích thước d (d-sized one-dimensional vectors), giả định đơn giản rằng những mã nhúng khác nhau tương đồng khía cạnh cùng một vị trí. Tương tự như ConvKB , nó ghép $h$,$r$ và $t$ vào một ma trận $ d \times 3 $ chiều. Đi qua một lớp tích chập với E 1 $\times$ 3 bộ lọc. Kết quả là d $\times$E ma trận trong cái thứ $i$-th giá trị của bất kỳ dòng duy nhất nào phụ thuộc vào $h[i]$, $r[i]$ vaf $t[i]$. Ma trận đi qua lớp bao bọc tách lớp bao bọc xử lý mỗi cột, nó nhận thông tin liên quan đến một khía cạnh của đầu vào sự việc . Lớp thứ hai với một lớp bao bọc thì được sử dụng để đưa ra bộ ba điểm số.

### 3.3.3 Recurrent Neural Networks (RNNs) (Mạng neural hồi qui)

Đây là những mô hình sử dụng một hoặc nhiều lớp hồi qui [[22]](https://github.com/hmthanh/thesis/blob/master/papers/22.%20Neural%20networks%20and%20physical%20systems%20with%20emergent%20collective%20computational%20abilities.pdf) để phân tích toàn bộ đường đi (sự kiện tuần tự) trích xuất từ tập dữ liệu huấn luyện. Thay vì chỉ xử lý các xự kiện một cách đọc lập tách biệt.

* **RSS** [ [19]](https://github.com/hmthanh/thesis/blob/master/papers/19_Learning_to_Exploit_Long-term_Relational_Dependencies_in_Knowledge_Graphs.pdf) dựa trên việc đề ý rằng RNNs cơ bản có thể không thích hợp cho LP, bởi vì chúng thực hiệp xử lý rõ ràng đường đi của các thực thể và quan hệ và khi dự đoán một sự kiện tail. Trong bước hiện tại chúng chỉ được thông qua quan hệ của nó, không phải head (đã được xét trước đó). Để khắc phục vấn đề này họ đề xuất Mạng Hồi Qui Nhảy  (RSNs):  vào mỗi bước, nếu đầu vào là một lên hệ (relation), trạng thái ẩn thì được cập nhật tái sử dụng những sự kiện head nữa. Điểm thực tế được tính khi dot product giữa vector đầu ra và mục tiêu nhúng. Trong quá trình huấn luyện, mô hình học đường dẫn quan hệ được xây dựng từ các sự kiện sử dụng kỹ thuật lấy mẫu biased random walk (**biased  random walk sampling**).
Nó sử dụng một hàm tối ưu đặc biệt nhờ đến định tương phản nhiễu (**type-based noise contrastive estimation**).
