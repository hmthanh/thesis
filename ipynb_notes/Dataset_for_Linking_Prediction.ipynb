### FB15k


là tập dữ liệu bao gồm trang web chứa ít nhất một siêu liên kết trỏ đến Wikipedia Tiếng Anh. Tập dữ liệu được lấy bằng cách lặp qua chỉ mục web của Google. Với mỗi trang trên wikipedia là một đại diện cho một thực thể và văn bản neo như một đề cập đến thực thể đó. 
Đây là dữ liệu gốc [Wikilinks](http://code.google.com/p/wiki-links/downloads/list)


FB15k là bộ dữ liệu phổ biến nhất làm chuẩn để dự đoán liên kết trong đồ thị phức cho đến nay. 
FB15k được tạo ra bằng cách chọn ra tất cả những thực thể của tập dữ liệu FreeBase với hơn 100 ghi chép và đặc trưng trong cơ sở dữ liệu Wikilinks, họ trích xuất ra tất cả các sự kiện liên quan đến chúng (vì thế bao gồm cả những node kế cận cấp thấp), ngoại trừ ngày tháng, danh từ riêng. Bon họ cũng chuyển mối quan hệ n-any bằng cách thống nhất thành một nhóm cạnh đi đôi (cliques of binary edges). thao tác này cực kỳ ảnh hưởng đến cấu trúc đồ thị và ngữ nghĩa (semantics)

### WN18

* Download : https://figshare.com/articles/WN18/11869548
* Dataset gốc : World Net (https://wordnet.princeton.edu)

WN18 (WordNet) đồng thời là tác giả của TransE [[8]](https://github.com/hmthanh/thesis/blob/master/papers/8_Translating_Embeddings_for_Modeling.pdf) được trích xuất từ World Net . Là một bản thể KG tri thức như thể là một từ điểm/ các từ đồng nghĩa để hỗ trợ xử lý ngôn ngữ tự nhiên và phân tích văn bản tự động. Trong thực thể World Net tương ứng với quá trình ngữ nghĩa hóa (word senses - synsets) và biểu diễn mỗi liên hệ (relations) của liên kết từ vựng. Để mà xây dựng WN18, tác giả sử dụng WordNet như là điểm khởi đầu và sau đó lọc qua toàn bộ các thực thể và mối quan hệ (relationships) với một ít đề cập.
