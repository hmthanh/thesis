## Quá trình embedded

Trong mô hình KBAT sử dụng TransE để nhúng tập dữ liệu

Ý tưởng của TransE là chuyển một cạnh thành một $\vec{vector}$ trong không gian, chuyển đổi **Translation** từ một entity (node) ~ $entity_{id}$ hoặc relation (cạnh) ~ $realtion_{id}$ trong KGs thành 1 vector không gian có k chiều $\vec{entity'}_k$ hoặc $\vec{relation'}_n$ (k tùy ý, trong mô hình KBAT k bằng 50) sao cho

$$
\vec{entity'_{head}} + \overrightarrow{relation'} = \vec{entity'_{tail}}
$$
