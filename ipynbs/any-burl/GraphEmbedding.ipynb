Đầu những năm 2000, các thuật toán nhúng đồ thị chủ yếu được thiết kế để giảm tính số chiều cao của dữ liệu không liên quan bằng cách giả sử dữ liệu nằm trong một đa tạp chiều thấp. Với một tập hợp các đặc trưng dữ liệu chiều cao không liên quan, biểu đồ tương tự được xây dựng dựa trên tính tương tự của đặc trưng cặp. Sau đó, mỗi **node** trong biểu đồ được nhúng vào một không gian chiều thấp nơi các **node** được kết nối gần nhau hơn. Ví dụ về dòng nghiên cứu này được giới thiệu trong Sec. 4.1. Một mặt, một số trong số chúng tập trung vào việc biểu diễn một phần của biểu đồ (ví dụ: **node**, cạnh, cấu trúc con) (Hình 1 (b) -1 (d)) dưới dạng một vectơ. Mặt khác, cũng có một số công việc tập trung vào việc nhúng toàn bộ biểu đồ dưới dạng một vectơ (Hình 1 (e)) cho các ứng dụng mức biểu đồ. **Graph kernels**(Phần 4.4) thường được thiết kế để đáp ứng nhu cầu này.

Cụ thể, nhúng đồ thị nhằm thể hiện một đồ thị dưới dạng các vectơ chiều thấp trong khi các cấu trúc biểu đồ được bảo tồn. Một mặt, phân tích đồ thị nhằm khai thác thông tin hữu ích từ dữ liệu biểu đồ. Việc nhúng đồ thị nằm trong sự chồng chéo của hai vấn đề và tập trung vào việc học các biểu diễn chiều thấp. Lưu ý rằng chúng tôi phân biệt học biểu diễn đồ thị và nhúng đồ thị trong khảo sát này. Mọi thứ nguyên biểu thị khoảng cách đo đạc của một nút với các nút khác trong biểu đồ.

Những thách thức của việc nhúng đồ thị phụ thuộc vào **problem setting** (cài đặt vấn đề), bao gồm đầu vào nhúng và đầu ra nhúng. Các loại đầu vào nhúng khác nhau mang thông tin khác nhau được bảo tồn trong không gian nhúng và do đó đặt ra những thách thức khác nhau đối với vấn đề nhúng đồ thị. Ví dụ, khi nhúng một biểu đồ chỉ với thông tin cấu trúc, các kết nối giữa các nút là mục tiêu cần được bảo tồn. Tuy nhiên, đối với biểu đồ có nhãn nút hoặc thông tin thuộc tính, thông tin phụ trợ cung cấp thuộc tính biểu đồ từ các phối cảnh khác và do đó cũng có thể được xem xét trong quá trình nhúng. Không giống như **embedding input** được đưa ra và cố định, đầu ra nhúng **embedding output** được điều khiển theo nhiệm vụ. Ví dụ, loại đầu ra nhúng phổ biến nhất là nhúng **node** đại diện cho các nút đóng như các vectơ tương tự. Tuy nhiên, trong một số trường hợp, các tác vụ có thể liên quan đến độ chi tiết cao hơn của đồ thị, ví dụ: cặp nút, đồ thị con, toàn bộ đồ thị. Do đó, thách thức đầu tiên về sản lượng nhúng là tìm ra loại đầu ra nhúng phù hợp cho ứng dụng quan tâm. Các mức độ chi tiết đầu ra khác nhau có các tiêu chí khác nhau để nhúng vào một đầu tốt và một số thách thức khác nhau. Ví dụ, một nút nhúng tốt bảo tồn sự tương tự với các nút lân cận của nó trong không gian nhúng. ### 2.1 Notation and Definition
Các mô tả chi tiết về các ký hiệu được sử dụng trong khảo sát này có thể được tìm thấy trong Bảng 1

*Definition 1 :* Một đồ thị là $G = (V, E)$, trong đó $v \in V$ là một node và $e \in E$ là một cạnh. $G$ được liên kết với hàm ánh xạ loại node $f_v: V \to T^v$ và hàm ánh xạ loại cạnh: $f_e: E \to T^e$

$T^v$ và $T^e$ lần lượt là tập hợp các loại node và loại cạnh. Mỗi nút $v_i \in V$ thuộc về một loại cụ thể, tức là, $f_v(vi) \in T^v$. Tương tự, đối với $e_{ij} \in E, f_e (eij) \in T^e$.

*Definition 2* Đồ thị đồng nhất $G_{homo} = (V, E)$ là đồ thị trong đó $\mid T^v \mid = \mid T^e \mid = 1$. Tất cả các node trong $G$ thuộc về một loại duy nhất và tất cả các cạnh thuộc về một loại duy nhất.

*Definition 3* Một đồ thị không đồng nhất $G_{hete} = (V, E)$ là một đồ thị trong đó $\mid T^v \mid > 1$ và / hoặc $\mid T^e \mid > 1$.

*Definition 4* Một đồ thị tri thức $G_{know} = (V, E)$ là một đồ thị có hướng có các nút là các thực thể (*entities*) và các cạnh (*edges*) là các sự kiện gồm bộ ba *subject-property-object*. Mỗi cạnh của biểu mẫu (*head entity, relation, tail entity*) (ký hiệu là $\langle h, r, t \rangle$) biểu thị mối quan hệ của $r$ từ thực thể $h$ đến thực thể $t$.

$h, t \in V$ là các thực thể và $r \in E$ là quan hệ. Trong khảo sát này, chúng tôi gọi $\langle h, r, t \rangle$ một bộ ba đồ thị tri thức. Ví dụ, trong Hình 3, có hai bộ ba: $\langle Alice, isF riendOf, Bob \rangle$ và $\langle Bob, isSupervisorOf, Chris \rangle$. Lưu ý rằng các thực thể và quan hệ trong đồ thị tri thức thường có các loại khác nhau [14 Translating on Hyperplanes](../../graph_embedding_paper/Embedding_by_Translating_on_Hyperplanes.pdf), [15 Learning Entity and Relation Embeddings](../../graph_embedding_paper/Learning_Entity_and_Relation_Embeddings.pdf). Do đó, biểu đồ tri thức có thể được xem như là một ví dụ của biểu đồ không đồng nhất.

Các biện pháp lân cận sau đây thường được áp dụng để định lượng thuộc tính đồ thị sẽ được bảo tồn trong không gian nhúng. Khoảng cách thứ tự đầu tiên là sự tương tự cặp đôi cục bộ giữa chỉ các nút được kết nối bởi các cạnh. Nó so sánh cường độ kết nối trực tiếp giữa một cặp nút. Chính thức

***Definition 5:*** Khoảng cách bậc nhất giữa nút ***first-order proximity*** $v_i$ và nút $v_j$ là trọng số của cạnh $e_{ij}$, tức là, $A_{i, j}$

Hai nút giống nhau hơn nếu chúng được kết nối bởi một cạnh có trọng lượng lớn hơn. Suy ra khoảng cách bậc nhất giữa nút $v_i$ và $v_j$ là $s^{(1)}_{ij}$, chúng ta có $s^{(1)}_{ij} = A_{i, j}$. Đặt $s^{(1)}_{i} = \begin{bmatrix} s^{(1)}_{i1}, s^{(1)}_{i2}, \dots, s^{(1)}_{i \mid V \mid} \end{bmatrix}$ biểu thị khoảng cách bậc nhất giữa vi và các nút khác. Lấy biểu đồ trong hình 1 (a) làm ví dụ, thứ tự đầu tiên giữa $v_1$ và $v_2$ là trọng số của cạnh $e_{12}$, ký hiệu là $s^{(1)}_{12} = 1.2$. Và $s^{(1)}_1$ ghi lại trọng số của các cạnh kết nối $v_1$ và các nút khác trong biểu đồ, tức là, $s^{(1)}_{1} = \begin{bmatrix}  0, 1.2, 1.5, 0, 0, 0, 0, 0, 0 \end{bmatrix} $.

Khoảng cách thứ hai so sánh sự giống nhau của các nút cấu trúc lân cận. Các vùng lân cận hai nút càng giống nhau, giá trị lân cận bậc hai giữa chúng càng lớn. Chính thức

Khoảng cách thứ hai so sánh sự giống nhau của các nút cấu trúc lân cận. Các vùng lân cận hai nút càng giống nhau, giá trị lân cận bậc hai giữa chúng càng lớn. Chính thức

***Definition 6:*** Khoảng cách thứ hai $s^{(2)}_{ij}$ ở giữa nút $v_i$ và $v_j$ là sự tương đồng giữa $v^{\prime}_i$ khu $s^{(1)}_i$ và $v^{\prime}_j$ ở $s$ khu $s^{(1)}_j$.

Một lần nữa, lấy hình 1 (a) làm ví dụ: $s^{(2)}_{12}$ là điểm tương đồng giữa $s^{(1)}_{1}$ và $s^{(1)}_{2}$. Như đã giới thiệu trước, $s^{(1)}_1 = \begin{bmatrix} 0, 1.2, 1.5, 0, 0, 0, 0, 0, 0 \end{bmatrix}$ và $s^{(1)}_2 = \begin{bmatrix} 1.2, 0, 0.8, 0, 0, 0, 0 , 0, 0 \end{bmatrix}$. Chúng ta hãy xem xét các điểm tương đồng cosine $s^{(2)}_{12} = cosine (s^{(1)}_1, s^{(1)}_2) = 0.43$ và $s^{(2)}_{15} = cosine(s^{(1)}_1, s^{(1)}_5) = 0$. Chúng ta có thể thấy rằng khoảng cách bậc hai giữa $v_1$ và $v_5$ bằng $0$ vì $v_1$ và $v_5$ không chia sẻ bất kỳ hàng xóm $1$ hop phổ biến nào. $v_1$ và $v_2$ chia sẻ một hàng xóm chung $v_3$, do đó khoảng cách thứ hai $s^{(2)}_{12} của chúng lớn hơn 0

Sự *higher-order proximity* (gần gũi bậc cao hơn_ có thể được định nghĩa tương tự. Ví dụ, khoảng cách thứ tự $k-th$ giữa nút $v_i$ và $v_j$ là sự tương đồng giữa $s^{(k 1)}_i$ và $s^{(k 1)}_j$. Lưu ý rằng đôi khi các giá trị gần đúng bậc cao hơn cũng được xác định bằng cách sử dụng một số số liệu khác, ví dụ: Katz Index, RootR PageRank, Adamic Adar, v.v. [11]

Đối với phần sau, các tài liệu được xử lý thêm để trích xuất các vectơ đặc trưng bằng cách sử dụng các kỹ thuật như bag-of-words [58], mô hình chủ đề [59], [60] hoặc xử lý từ Word như một loại nút [61]. Các loại đặc trưng nút khác, chẳng hạn như đặc trưng hình ảnh [33], cũng có thể. Các đặc trưng của nút tăng cường hiệu suất nhúng biểu đồ bằng cách cung cấp thông tin phong phú và không có cấu trúc, có sẵn trong nhiều đồ thị trong thế giới thực. Hơn nữa, nó làm cho đồ thị quy nạp có thể nhúng [62].

***Information propagation*** Truyền thông tin: Một ví dụ về việc truyền bá thông tin là truyền thông trực tuyến trên Twitter. Trong [63], được đưa ra biểu đồ dữ liệu $G = (V, E)$, biểu đồ phân tầng $G^c = (V^c, E^c)$ được xây dựng cho mỗi tầng $c$, trong đó $V^c$ là các nút đã áp dụng $c$ và $E^c$ là các cạnh với cả hai đầu trong $V^c$. Sau đó, họ nhúng $G^c$ để dự đoán sự gia tăng của kích thước tầng. Khác nhau, [64] nhằm mục đích nhúng người dùng và thông tin nội dung, sao cho sự giống nhau giữa việc nhúng của họ cho thấy xác suất khuếch tán. Topo-LSTM [65] coi một tầng không chỉ đơn thuần là một chuỗi các nút, mà là một đồ thị chu kỳ có hướng động để nhúng.

***Knowledge base*** Cơ sở tri thức: Các cơ sở tri thức phổ biến bao gồm Wikipedia [66], Freebase [37], YAGO [67], DBpedia [68], v.v ... Lấy Wikipedia làm ví dụ, các khái niệm là các thực thể được đề xuất bởi người dùng và văn bản là bài viết liên quan đến thực thể. [66] sử dụng nền tảng kiến thức để tìm hiểu biểu đồ tri thức xã hội từ mạng xã hội bằng cách liên kết mỗi người dùng mạng xã hội với một tập hợp các khái niệm kiến thức nhất định. [69] thể hiện các truy vấn và tài liệu trong không gian thực thể (được cung cấp bởi một cơ sở tri thức) để công cụ tìm kiếm hàn lâm có thể hiểu ý nghĩa của các khái niệm nghiên cứu trong các truy vấn

Các loại thông tin phụ trợ khác bao gồm dữ liệu đăng nhập người dùng (vị trí người dùng) [70], danh sách xếp hạng tùy chọn mục người dùng [71], v.v ... Lưu ý rằng thông tin phụ trợ không chỉ giới hạn ở một loại. Ví dụ: [62] và [72] xem xét cả thông tin tính năng nhãn và nút. [73] sử dụng nội dung nút và nhãn để hỗ trợ quá trình nhúng biểu đồ.

Thách thức: Làm thế nào để kết hợp thông tin phong phú và không có cấu trúc sao cho các nhúng được học vừa thể hiện cấu trúc tôpô và phân biệt đối xử về mặt thông tin phụ trợ? Thông tin phụ trợ giúp xác định độ tương tự của nút ngoài thông tin cấu trúc biểu đồ. Những thách thức của việc nhúng đồ thị với thông tin phụ trợ là làm thế nào để kết hợp hai nguồn thông tin này để xác định độ tương tự của nút cần được bảo tồn.
